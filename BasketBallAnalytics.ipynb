### Now, we have created data matrices needed ! Let's move to visualisation and insights!

```r
myplot <- function(data,pos,rows=1:10){
  Data <- data[rows,,drop=F]
  matplot(t(Data),type="b", pch=15:18,col=c(1:4,6),main="Basketball Players Analysis")
  legend(pos,inset=0.01,legend=Players[rows],col=c(1:4,6),pch=15:18,horiz = F, bg='white',cex = 0.75)
}
```

## Player Style excluding Free Throws

```r
# Points per field goals
myplot((Points-FreeThrows)/FieldGoals,"topleft")
```

### Because we have excluded free throws, this plot shows us the true representation of player style change. We can verify that this is the case because all the marks without exception on this plot are between 2 and 3. That is because Field Goalscan only be for either 2 points or 3 points.

### Insights:
<br> 1. You can see how players' preference for 2 or 3 point shots changes throughout their career. We can see that almost all players in this dataset experiment with their style throughout their careers. Perhaps, the most drastic change in style has been experienced by <b>Joe Johnson</b>.</br>

<br> 2. There is one exception. You can see that one player has not changed his style at all - almost always scoring only 2-pointers. Who is this mystery player? It's <b>Dwight Howard!</b> Now that explains a lot. The reason that Dwight Howard's Field Goal accuracy is so good is because he almost always scores 2-pointers only. That means he can be close to the basket or even in contact with it. Free throws, on the other hand require the player to stand 15ft (4.57m) away from the hoop. That's probably why Dwight Howard's Free Throw Accuracy is poor.</br>
